```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from(['a','b','c','d'])
```

```python
G.add_edges_from([('a','b'),('a','c'),('b','c'),('c','d')])
```

```python
nx.draw(G,with_labels=True,node_color="pink",node_size=1000,font_size=20,font_color="black")
```

```python
dict(G.degree())
```

```python
nx.to_numpy_matrix(G)
```

```python
import pandas as pd
```

```python
df =pd.read_csv('HomoSapiens_binary_hq.txt',nrows=100,sep='\t',usecols=['Gene_A','Gene_B'])
```

```python
df
```

```python
bio_graph=nx.from_pandas_edgelist(df,'Gene_A','Gene_B',create_using=nx.DiGraph)
```

```python
nx.draw(bio_graph,with_labels=True,node_color="pink",node_size=1000,font_size=20,font_color="black")
```

```python
import matplotlib.pyplot as plt
```

```python
degrees=dict(bio_graph.degree)
order=sorted(degrees.items(),key=lambda t :t[1])

x,y=zip(*order)
plt.plot(x,y)
plt.show()
```
